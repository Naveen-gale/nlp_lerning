```python
import pandas as pd
from sklearn.model_selection import train_test_split
import string
import re
import nltk
from nltk.corpus import stopwords

# Download NLTK data if not already available (to ensure cell runs independently)
try:
    nltk.data.find('corpora/stopwords')
except nltk.downloader.DownloadError:
    nltk.download('stopwords')
try:
    nltk.data.find('tokenizers/punkt') # 'punkt' is often needed with 'stopwords' for tokenization
except nltk.downloader.DownloadError:
    nltk.download('punkt')

# Re-load df to ensure 'emotion' column is present and data is correct
df = pd.read_csv('train.txt', sep=';', header=None, names=['text', 'emotion'])

# Re-apply text preprocessing steps (as per the original notebook)
df['text'] = df['text'].apply(lambda x: x.lower())

def remove_puch(text):
  return text.translate(str.maketrans('','',string.punctuation))
df['text'] = df['text'].apply(lambda x: remove_puch(x))

def remove_num(text):
    return re.sub(r'\\d+', '', text)
df['text'] = df['text'].apply(remove_num)

df["text"] = df["text"].apply(
    lambda text: re.sub(r'https?://\\S+|www\\.\\S+', '', text)
)

df["text"] = df["text"].apply(
    lambda text: re.sub(r'<.*?>', '', text)
)

def remove_emoji(text):
    new = ''
    for eoj in text:
        if eoj.isnumeric() or eoj.isalpha() or eoj == ' ':
            new += eoj
    return new
df['text'] = df['text'].apply(remove_emoji)

stop_words = set(stopwords.words('english'))
df["text"] = df["text"].apply(
    lambda text: " ".join(
        word for word in text.split()
        if word.lower() not in stop_words
    )
)

# Now, perform the train-test split with the corrected df
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.33, random_state=42)
```
