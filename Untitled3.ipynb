```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python

df = pd.read_csv('train.txt', sep=';', header=None, names=['text', 'emotion'])
df.head()
```

```python
unik_emotion = df['emotion'].unique()
unik_number={}
i= 0
for eom in unik_emotion:
  unik_number(eom) = i
  i+=1

df['emotion_number'] = df['emotion'].map(unik_number)
```

```python
df['text']=df['text'].apply(lambda x: x.lower())
```

```python
import string
def remove_puch(text):
  return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['text']=df['text'].apply(lambda x: remove_puch(x))
```

```python
import re

def remove_num(text):
    return re.sub(r'\d+', '', text)
```

```python
df['text']= df['text'].apply(remove_num)
```

```python
df.head()
```

```python
df["text"] = df["text"].apply(
    lambda text: re.sub(r'https?://\S+|www\.\S+', '', text)
)
```

```python
df["text"] = df["text"].apply(
    lambda text: re.sub(r'<.*?>', '', text)
)
```

```python
def remove_emoji(text):
    new = ''
    for eoj in text:
        if eoj.isnumeric() or eoj.isalpha() or eoj == ' ':
            new += eoj
    return new
```

```python
df['text']= df['text'].apply(remove_emoji)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize  import word_tokenize
```

```python
nltk.download('stopwords')
nltk.download('punkt')
stop_words = set(stopwords.words('english'))
```

```python
df.loc[1]['text']
```

```python
df["text"] = df["text"].apply(
    lambda text: " ".join(
        word for word in text.split()
        if word.lower() not in stop_words
    )
)
```
